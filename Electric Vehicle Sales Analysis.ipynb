 <center><h3 style="background-color:#6030ab;"><i><font color=" #fffff">Clean And Prepare The Data <hr></font><i></h3></center>

```python
#Checking if the Datatypes are correct or not
df.info()
```

```python
 #The column Year is given as Float, but we should convert it to int.
df['Year'] = df['Year'].astype(int)
```

```python
#The column Date is given as Object, but it should be in Datetime format
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python
#Extract Month and Date
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
```

```python
 #Similarly converting other columns to their relevant datatypes.
categorial_columns = ['Month_Name', 'State','Vehicle_Class','Vehicle_Category', 'Vehicle_Type']
df[categorial_columns] = df[categorial_columns].astype('category')
```

```python
 #Checking once again
print(' Now everything is well organised')
print(df.info())
```

 <center><h3 style="background-color:#6030ab;"><i><font color=" #fffff">Business Problem Analysis <hr></font><i></h3></center>

```python
#1. Top 10 States by total EV Sales
state_sales = df.groupby('State')['EV_Sales_Quantity'].sum().sort_values(ascending=False).head(10)
state_sales.plot(kind='barh', title='Top 10 States By EV Sales', color='skyblue')
plt.xlabel('EV Sales Quantity')
plt.gca().invert_yaxis()
```

```python
# 2. Most preferred vehicle type
types_sales = df.groupby('Vehicle_Type')['EV_Sales_Quantity'].sum().sort_values(ascending=False)
types_sales.plot(kind='bar', title='EV Sales By Vehicle Type', color='orange')
plt.ylabel('EV Sales Quantity')
plt.xticks(rotation=45)
```

```python
# 3. EV growth trend in top 5 states
top_states = state_sales.index.tolist()
for state in top_states:
    state_data=df[df['State']== state].groupby('Year')['EV_Sales_Quantity'].sum()
    plt.plot(state_data.index, state_data.values,label=state)

plt.title('EV Sales Growth in Top 5 States')
plt.xlabel('Year')
plt.ylabel('EV Sales Quantity')
plt.legend()
plt.grid(True)
   
```

```python
# 4. Year-wise EV sales
year_sales = df.groupby('Year')['EV_Sales_Quantity'].sum()
year_sales.plot(kind='line', marker='o', title='EV Sales in  India Over Years', color='green')
plt.ylabel('Total EV Sales')
plt.grid(True)
```

```python
# 5. Month-wise EV sales
month_sales = df.groupby('Month_Name')['EV_Sales_Quantity'].sum().sort_values(ascending=False)
month_sales.plot(kind='bar', title='Month-Wise Sales',color='purple')
plt.ylabel('EV Sales Quantity')
plt.xticks(rotation=45)
```

```python
# 6. Vehicle category share
cat_sales = df.groupby('Vehicle_Category')['EV_Sales_Quantity'].sum()
cat_sales.plot(kind='pie', autopct='%1.1f%%', startangle=90, title='Vehicle Category Share')
plt.ylabel('')
```

```python
# 7. Top 10 vehicle classes
vc_sales= df.groupby('Vehicle_Class')['EV_Sales_Quantity'].sum().sort_values(ascending=False).head(10)
vc_sales.plot(kind='bar', title='Top 10 Vehicle Class by EV Sales', color='teal')
plt.ylabel('EV Sales Quantity')
plt.xticks(rotation=45)
```

```python
# 8. Personal vs Shared Vehicle Usage
df['Usage_Type'] = df['Vehicle_Type'].apply(lambda x: 'Personal' if 'Personal' in x else ('Shared' if 'Shared' in x else 'Others'))
usage_sales = df.groupby('Usage_Type')['EV_Sales_Quantity'].sum()
usage_sales.plot(kind='bar', title="Personal vs Shared Usage", color=['green', 'red', 'gray'])
plt.ylabel("EV Sales Quantity")
```
